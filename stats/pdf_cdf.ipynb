```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
data=sns.load_dataset('iris')
sns.FacetGrid(data=data,hue='species',height=6).map(sns.displot,'petal_length').add_legend()
```

```python
#pdfs give us the probability of the random varaible, and is used to try and understand whtehr or not a hue will occur or not 
#when pdfs are spearate, they can seggregate the data better vs if they are overlapping 
#we use them in place of histogram 
```

```python
##cdf 
#gives us the percentage(y axis) of all the observations that are at a certain value(x axis)
```

```python
##how to build a pdf and cdf 
#pdf-freq of a random var/total random vars, gives us p of a certain point
#cdf-gives us p of a certain range <=
```

```python
counts,bin_edges=np.histogram(data['petal_length'])
pdf=counts/sum(counts)
print(pdf)
cdf=np.cumsum(pdf)
print(cdf)
```

```python
plt.plot(bin_edges[:-1],pdf)
plt.plot(bin_edges[:-1],cdf)
```
