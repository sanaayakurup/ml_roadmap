```python
##MISSING DATA 
#missing data
total = data.isnull().sum().sort_values(ascending=False)
percent = (data.isnull().sum()/data.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
cols_to_remove=missing_data[missing_data['Total']>1].index
cols_to_remove
```

```python
missing_data.head(20)
```

```python
data.drop(columns=cols_to_remove,inplace=True) #columns removed 
```

```python
a[a>0.4]['SalePrice'].sort_values(ascending=False)[0:14] #important positively correlated vars 
```

```python
##remove electrical 
data.drop(data[data['Electrical'].isnull()].index[0],inplace=True) #one empty row removed 
```

```python
data.sort_values(by = 'GrLivArea', ascending = False)[:2] #biggest 2 vals 
data = data.drop(data[data['Id'] == 1299].index)
data = data.drop(data[data['Id'] == 524].index)
```

```python
##normality 
from scipy.stats import norm
sns.distplot(data['SalePrice'],fit=norm)
data['SalePrice']=np.log(data['SalePrice'])
```

```python
from scipy import stats
sns.distplot(data['SalePrice']) #now it looks normal 
```

```python
sns.distplot(data['SalePrice'], fit=norm);
fig = plt.figure()
res = stats.probplot(data['SalePrice'], plot=plt)
```

```python
#Done! Let's check what's going on with 'GrLivArea'.
sns.distplot(data['GrLivArea'])
fig = plt.figure()
res = stats.probplot(data['SalePrice'], plot=plt)
```

```python
sns.distplot(data['GrLivArea'])
fig = plt.figure()
res = stats.probplot(data['SalePrice'], plot=plt)
```

```python
data['GrLivArea']=np.log(data['GrLivArea'])
```

```python
#histogram and normal probability plot
sns.distplot(data['TotalBsmtSF'], fit=norm);
fig = plt.figure()
res = stats.probplot(data['TotalBsmtSF'], plot=plt)
```

```python
#we have 0 values, thus log of this is not possible 
data['HasBsmt']=pd.Series(len(data['TotalBsmtSF']), index=data.index)
data['HasBsmt'] =0 
```

```python
data.loc[data['TotalBsmtSF']>0,'HasBsmt']=1
```

```python
data.loc[data['HasBsmt']==1,'TotalBsmtSF'] = np.log(data['TotalBsmtSF'])
```

```python
sns.distplot(data[data['TotalBsmtSF']>0]['TotalBsmtSF'], fit=norm);
fig = plt.figure()
res = stats.probplot(data[data['TotalBsmtSF']>0]['TotalBsmtSF'], plot=plt)
```
