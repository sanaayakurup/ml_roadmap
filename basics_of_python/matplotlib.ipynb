```python
##stack charts 
minutes = [1, 2, 3, 4, 5, 6, 7, 8, 9]

player1 = [8, 6, 5, 5, 4, 2, 1, 1, 0]
player2 = [0, 1, 2, 2, 2, 4, 4, 4, 4]
player3 = [0, 1, 1, 1, 2, 2, 3, 3, 4]
labels=['player1','p2','p3']
#stack chart-indi contri+total contri in a single chartt
plt.stackplot(minutes,player1,player2,player3,labels=labels)
plt.xlabel('Players contri')
plt.ylabel('Minutes')
plt.title('total goals')
plt.legend()
```

```python
#linbe chart reading
#https://www.youtube.com/watch?v=x0Uguu7gqgk
data=pd.read_csv('https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/05-Fill_Betweens/data.csv')
data.head()
```

```python
plt.plot(data['Age'],data['All_Devs'],label='all_dev')
plt.plot(data['Age'],data['Python'],label='p_Dev',linestyle='--')
overall_median=57287
plt.fill_between(data['Age'],data['Python'],overall_median,where=(data['Python']>overall_median),alpha=0.25,interpolate=True,label='above avg')
plt.fill_between(data['Age'],data['Python'],overall_median,where=(data['Python']<=overall_median),alpha=0.25,interpolate=True,label='below avg')
plt.legend()
plt.xlabel('Age')
plt.ylabel('Sal')
plt.title('Age vs median sal')
```

```python
#subplots https://www.youtube.com/watch?v=XFZRVnP-MTU&list=PL-osiE80TeTvipOqomVEeZ1HRrcEvtZB_&index=10
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/10-Subplots/data.csv')
data.head()
```

```python
fig,ax=plt.subplots() #our ax is set to 1,1 by default. we need to pass our rows and columns for multiple rows and cols 
ax.plot(data['Age'],data['All_Devs'],label='All devs')
ax.plot(data['Age'],data['Python'],label='Py dev')
ax.plot(data['Age'],data['JavaScript'],label='Java dev')
ax.legend()
ax.set_xlabel('age') 
ax.set_ylabel('sal')
```

```python
fig,(ax1,ax2)=plt.subplots(2,1) #two rows 1 col
ax1.plot(data['Age'],data['All_Devs'],label='All devs')
ax2.plot(data['Age'],data['JavaScript'],label='J dev')
ax2.set_xlabel('Age')
```
