```python
import seaborn as sns #https://www.youtube.com/watch?v=K_8u_AJ75RI&list=PLKnIA16_RmvbB1bFGjvS6a8T0mnqawejo&index=1
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
data=sns.load_dataset('tips')
data.head()
```

```python
#plot scatter using total bill and tip in pyplot 
plt.scatter(data['total_bill'],data['tip'])
plt.xlabel('Bill')
plt.ylabel('Tip')
plt.title('Bill vs tip')
```

```python
sns.scatterplot(data['total_bill'],data['tip'],hue=data['smoker']) #using scatter plot 
```

```python
sns.relplot(data['total_bill'],data['tip'],hue=data['smoker'],style=data['sex']) #categorical data will be in stype and hue datab
```

```python
#the size param needs continous data 
sns.relplot(data['total_bill'],data['tip'],hue=data['smoker'],style=data['sex'],size=data['size'])
```

```python
#heatmaps 
flights=sns.load_dataset('flights')
flights.head()
```

```python
fight_pivot=flights.pivot_table(index='year',values='passengers',columns='month',aggfunc='sum')
sns.heatmap(fight_pivot,linewidth=0.4)
```

```python
#jointplots https://www.youtube.com/watch?v=yV4pSrX_2h4&list=PLKnIA16_RmvbB1bFGjvS6a8T0mnqawejo&index=8
tips=sns.load_dataset('tips')
sns.jointplot(x='total_bill',y='tip', data=tips) #plots scatter+hist 
```

```python
##pairplot
data=sns.load_dataset('iris')
sns.pairplot(data,hue='species')
```

```python
##distplot- distribution plots -three graph-hist, kde, rugplot
data=sns.load_dataset('titanic')
data.head()
data.isna().sum()
data.age.fillna(data.age.mean(),inplace=True)
data.isna().sum()
```

```python
sns.distplot(data['age'])
```

```python
#plot only hist 
sns.distplot(data['age'],kde=False,bins=10)
```

```python
sns.distplot(data['age'],hist=False)
```

```python
sns.distplot(data['age'],rug=True,hist=False,kde=False)
```
